### 1.2 Adding another plot

```python
x = [0,1,2,3,4,5,6]
y_1 = [0,2,4,6,8,10,12]
y_2 = [0,3,6,9,12,15,18]

plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x,y_2, c='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0,1,2,3,4,5,6]
y_1 = [0,2,4,6,8,10,12]
y_2 = [0,3,6,9,12,15,18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x,y_2, c='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
            c='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
import numpy as np

x=[0]

for a in range (1,51):
    x += [a]
    
print(x)
```

```python
#np.negative (for - value in EXP)
y = np.sin(x)*np.exp(np.negative(x)/10)
plt.ylim([-1,1])

x=np.linspace(0,51,300)

plt.plot(x,y)
plt.xlabel('x', color='darkblue', fontsize=15)
plt.ylabel('y', color='darkblue', fontsize=15)
plt.grid(alpha=.25)
plt.title('$y=e^{-x/10}\sin(x)$', fontsize=20)

plt.tight_layout()
plt.show()
```
